### Total number of schools
There are 15 schools in total in the dataset provided for this project

```python
#district summary

total_schools = school_data_complete["school_name"].nunique()
student_sum = school_data_complete["Student ID"].nunique()
total_budget = "${:,.2f}".format(school_data["budget"].sum())

average_maths_score = round((school_data_complete["math_score"].mean()), 2)
average_read_score = round((school_data_complete["reading_score"].mean()), 2)

goodmathsers = school_data_complete[school_data_complete["math_score"] >= 70]
pass_maths = round((len(goodmathsers)/student_sum * 100), 2)

goodreaders = school_data_complete[school_data_complete["reading_score"] >= 70]
pass_read = round((len(goodreaders)/student_sum * 100), 2)

goodboth = school_data_complete[(school_data_complete["reading_score"] >= 70) & (school_data_complete["math_score"] >= 70)]
pass_both = round((len(goodboth)/student_sum * 100), 2)

district_df = pd.DataFrame({
    "Total Schools": [total_schools],
    "Total Students": [student_sum],
    "Total Budget": [total_budget], 
    "Average Maths Score": [average_maths_score],
    "Average Reading Score": [average_read_score],
    "% Passing Maths": [pass_maths],
    "% Passing Reading": [pass_read],
    "% Overall Passing": [pass_both],
})
district_df
```

```python
x = school_data_complete.groupby("school_name")
y = x["reading_score"].mean()
z= y.mean()
z
```

```python
#school summary

school_sum = school_data_complete.groupby("school_name").agg({
                                                                "type": lambda column: column.unique(), 
                                                                "size": lambda column: column.unique(), 
                                                                "budget": lambda column: column.unique(),
                                                                "reading_score": "mean", 
                                                                "math_score": "mean"})
#means = school_data_complete.groupby(["school_name"]).mean()
#school_sum = means[["size", "budget", "math_score", "reading_score"]]

#rename the existing columns
school_sum = school_sum.rename(columns={
                                "type": "Type",
                                "size": "Total Students",
                                "budget": "Total School Budget",
                                "math_score": "Average Math Score",
                                "reading_score": "Average Reading Score"})

#calculate averages
school_sum["Per Student Budget"] = (school_sum["Total School Budget"]/school_sum["Total Students"])

x = goodmathsers.groupby(["school_name"])
school_sum["% Passing Math"] = round((x.count()["math_score"]/school_sum["Total Students"] * 100), 2)

y = goodreaders.groupby(["school_name"])
school_sum["% Passing Reading"] = round((y.count()["reading_score"]/school_sum["Total Students"] * 100), 2)

z = goodboth.groupby(["school_name"])
school_sum["% Overall Passing"] = round((z.count()["reading_score"]/school_sum["Total Students"] * 100), 2)



#change the order of columns:
school_sum = school_sum[[
      "Type",   
      "Total Students", 
      "Total School Budget", 
      "Per Student Budget", 
      "Average Math Score", 
      "Average Reading Score",
      "% Passing Math",
      "% Passing Reading",
      "% Overall Passing"
     ]]

#format numbers nicely
school_sum_dollar = round(school_sum, 2)
school_sum_dollar["Total School Budget"] = school_sum_dollar["Total School Budget"].map("${:,.2f}".format)
school_sum_dollar["Per Student Budget"] = school_sum_dollar["Per Student Budget"].map("${:,.2f}".format)
school_sum_dollar
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
# Create the bins in which Data will be held   
spend_bins = [0, 584.99, 629.99, 644.99, 680]

# Create the names for the five bins
spend_categs = ["<$585", "$585-630", "$630-645", "$645-680"]

#cut data into bins and summarise into new dataframe
school_sum["Spending Ranges (Per Student)"] = pd.cut(school_sum["Per Student Budget"], spend_bins, labels=spend_categs, include_lowest=True)
spend_summary = school_sum.groupby(["Spending Ranges (Per Student)"]).mean()

#trim back to relevant columns and display
spend_summary = spend_summary[[ 
      "Average Math Score", 
      "Average Reading Score",
      "% Passing Math",
      "% Passing Reading",
      "% Overall Passing"
     ]]
spend_summary = round(spend_summary, 2)
spend_summary 
```

## Scores by School Size

* Perform the same operations as above, based on school size.

```python
# Create the bins in which Data will be held   
size_bins = [0, 999, 1999, 5000]

# Create the names for the five bins
size_categs = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000) "]

#cut data into bins and summarise into new dataframe
school_sum["School Size"] = pd.cut(school_sum["Total Students"], size_bins, labels=size_categs, include_lowest=True)
size_summary = school_sum.groupby(["School Size"]).mean()

#trim back to relevant columns and display
size_summary = size_summary[[ 
      "Average Math Score", 
      "Average Reading Score",
      "% Passing Math",
      "% Passing Reading",
      "% Overall Passing"
     ]]
size_summary = round(size_summary, 2)
size_summary
```
